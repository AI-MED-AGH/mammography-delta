```python
import sys
sys.path.append("..")

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score

from preprocessing import build_preprocessor
from baseline_logreg import build_baseline_logreg
from result import evaluate_baseline_results
```

```python
df = pd.read_csv("../../simplified-CBIS-DDSM/features.csv")

TARGET_MAPPING = {
    "BENIGN": 0,
    "MALIGNANT": 1
}
X = df.drop(columns=["pathology"])
y = df["pathology"].map(TARGET_MAPPING)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

preprocessor = build_preprocessor(X_train)
model = build_baseline_logreg(preprocessor)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)[:, 1]

accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_proba)

print(f"Accuracy: {accuracy:.4f}")
print(f"F1-score: {f1:.4f}")
print(f"ROC-AUC: {roc_auc:.4f}")
```

```python
y_random = np.random.randint(0, 2, size=len(y_test))
y_random_proba = np.random.rand(len(y_test))

random_accuracy = accuracy_score(y_test, y_random)
random_f1 = f1_score(y_test, y_random)
random_roc_auc = roc_auc_score(y_test, y_random_proba)

print(f"Accuracy: {random_accuracy:.4f}")
print(f"F1-score: {random_f1:.4f}")
print(f"ROC-AUC: {random_roc_auc:.4f}")
```

```python
results = pd.DataFrame({
    "Model": ["Random", "Logistic Regression (Baseline)"],
    "Accuracy": [random_accuracy, accuracy],
    "F1-score": [random_f1, f1],
    "ROC-AUC": [random_roc_auc, roc_auc]
})

results
```

```python
evaluate_baseline_results(results)
```
