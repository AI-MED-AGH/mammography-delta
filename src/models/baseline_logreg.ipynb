```python
import sys
sys.path.append("..")

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from preprocessing import build_preprocessor
from result import evaluate_baseline_results
```

```python
# Build basic Logistic Regression model
def build_baseline_logreg(preprocessor):
    return Pipeline(steps=[
        ("preprocessing", preprocessor),
        ("classifier", LogisticRegression(
            max_iter=1000,
            random_state=42
        ))
    ])
```

```python
# Read features from features.csv file, drop unnecessary columns and split into X and y
df = pd.read_csv("../../output/features.csv")
X = df.drop(columns=["pathology", "assessment", "patient_id", "id"])
y = df["pathology"]
```

```python
# Train basic Logistic Regression model and evaluate it
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

preprocessor = build_preprocessor(X_train)
model = build_baseline_logreg(preprocessor)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)[:, 1]

accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_proba)

print(f"Accuracy: {accuracy:.4f}")
print(f"F1-score: {f1:.4f}")
print(f"ROC-AUC: {roc_auc:.4f}")
```

```python
# Train model on random data
y_random = np.random.randint(0, 2, size=len(y_test))
y_random_proba = np.random.rand(len(y_test))

random_accuracy = accuracy_score(y_test, y_random)
random_f1 = f1_score(y_test, y_random)
random_roc_auc = roc_auc_score(y_test, y_random_proba)

print(f"Accuracy: {random_accuracy:.4f}")
print(f"F1-score: {random_f1:.4f}")
print(f"ROC-AUC: {random_roc_auc:.4f}")
```

```python
# Compare Random and Logistic Regression results
results = pd.DataFrame({
    "Model": ["Random", "Logistic Regression (Baseline)"],
    "Accuracy": [random_accuracy, accuracy],
    "F1-score": [random_f1, f1],
    "ROC-AUC": [random_roc_auc, roc_auc]
})

results
```

```python
# Automatically evaluates whether baseline results indicate a meaningful predictive signal.
evaluate_baseline_results(results)
```
