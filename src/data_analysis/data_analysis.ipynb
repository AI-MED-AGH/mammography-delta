```python
import pandas
import os

img_folder = 'images/'
CSV = '../../labels.csv'
data = pandas.read_csv(CSV, delimiter=',')

# Check missing values
missing_values = data.isnull().sum().sum()
if missing_values > 0:
    print(f"Found {missing_values} missing values in CSV.")
else:
    print("No missing values in CSV.")

if 'pathology' in data.columns:
        print("\nClass distribution (pathology):")
        print(data['pathology'].value_counts())

total_records = len(data)
print(f"Total records in CSV: {total_records}")

# Check if masks exist
data['mask_filename'] = data['id'].astype(str) + '.png'
data['mask_path'] = data['mask_filename'].apply(lambda x: os.path.join(img_folder, x))

# Check if file exists
data['mask_exists'] = data['mask_path'].apply(os.path.exists)

masks_count = data['mask_exists'].sum()
missing_count = total_records - masks_count

print(f"\nBinary masks:")
print(f"Existing: {masks_count}")
print(f"Missing:  {missing_count}")
```

```python
# Count of assessment value for further analysis
assessment = data['assessment'].value_counts()
assessment.sort_index(inplace=True)
print(f"\nAssessment distribution: {assessment}")
```

```python
# Checking multiple masks of patients
patients_masks_count = data['patient_id'].value_counts()
multiple_patients_masks = patients_masks_count[patients_masks_count > 1]

single_patients_masks = patients_masks_count[patients_masks_count == 1]
```

```python
print(f"Number of actual patients: {patients_masks_count.count()}")
```

```python
print(f"Number of patients examined once: {single_patients_masks.count()}")
print("\nPatients examined once:")
single_patients_masks
```

```python
print(f"\nNumber of patients examined more than once: {multiple_patients_masks.count()}")
print("\nPatients examined more than once: ")
multiple_patients_masks
```
